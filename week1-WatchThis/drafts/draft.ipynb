```python
def merger(a, b):
    if pd.notna(a) and pd.notna(b): return (a + b) / 2
    elif pd.notna(a): return a
    elif pd.notna(b): return b
    else: return None
```

```python
df['merged_rating']=df.apply(lambda row: merger(row['imdb.rating.$numberDouble'],row['tomatoes.viewer.rating.$numberDouble']), axis=1)
```

```python
def merge_plot(full, part):
    if pd.isna(full): full = ""  
    if pd.isna(part): part = ""
    return f"{full}\nShortly: {part}".strip()
```

```python
df["merged_plot"] = df.apply(lambda row: merge_plot(row['fullplot'], row['plot']), axis=1)
df
```

```python
import requests
```

```python
hf_token = "***"
embedding_url="https://api-inference.huggingface.co/pipeline/feature-extraction/sentence-transformers/all-MiniLM-L6-v2"
```

```python
def generate_embedding(text:str) -> list[float]:
    response = requests.post(
        embedding_url,
        headers={"Authorization": f"Bearer {hf_token}"},
        json={"inputs": text})
    if response.status_code != 200:
        raise ValueError(f"Request failed with status code {response.status_code}: {response.text}")
    return response.json()
```
